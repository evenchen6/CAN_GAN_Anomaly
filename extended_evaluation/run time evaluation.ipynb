## Performance Analysis Evaluation

```python
# define time data path
import numpy as np
import matplotlib.pyplot as plt

run_in_tvm_cpu_size_0_path = './run_in_tvm_cpu_size_0.npz' # Single-core
run_in_tvm_quantified_cpu_size_0_path = './run_in_tvm(quantified)_cpu_size_0.npz' # Single-core(quantized)
run_in_tvm_cpu_size_4_path = './run_in_tvm_cpu_size_4.npz' # Multicore
run_in_tvm_quantified_cpu_size_4_path = './run_in_tvm(quantified)_cpu_size_4.npz' # Multicore(quantized)

run_in_tvm_cpu_size_0_time_data = np.load(run_in_tvm_cpu_size_0_path)["time_data"]
run_in_tvm_quantified_cpu_size_0_time_data = np.load(run_in_tvm_quantified_cpu_size_0_path)["time_data"]
run_in_tvm_cpu_size_4_time_data = np.load(run_in_tvm_cpu_size_4_path)["time_data"]
run_in_tvm_quantified_cpu_size_4_time_data = np.load(run_in_tvm_quantified_cpu_size_4_path)["time_data"]
```

### Single-core vs Single-core(quantized) （cpu num is 1）

```python
# performance evaluation
# Single-core
plt.style.use('seaborn-white')
plt.title('Execution time distribution of ACGAN-AC-RFC-Th model running on single core environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

plt.hist(run_in_tvm_cpu_size_0_time_data, bins=300)
```

```python
# performance evaluation
# Single-core(quantized)
plt.style.use('seaborn-white')
plt.title('Execution time distribution of ACGAN-AC-RFC-Th model(quantized) running on single core environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

plt.hist(run_in_tvm_quantified_cpu_size_0_time_data,bins=300)
```

```python
# Single-core vs Single-core(quantized)
plt.title('ACGAN-AC-RFC-Th model vs ACGAN-AC-RFC-Th model(quantized) on single core environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

x1 = run_in_tvm_cpu_size_0_time_data
x2 = run_in_tvm_quantified_cpu_size_0_time_data
kwargs1 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Single-core")
kwargs2 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Single-core(quantized)")
plt.hist(x1, **kwargs1)
plt.hist(x2, **kwargs2)
plt.legend() # show label
```

### Multicore vs Multicore(quantized) (cpu nums is 4)

```python
# performance evaluation
# Multicore
plt.style.use('seaborn-white')
plt.title('Execution time distribution of ACGAN-AC-RFC-Th model running on multicore environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

plt.hist(run_in_tvm_cpu_size_4_time_data, bins=300)
```

```python
# performance evaluation
# Multicore(quantized)
plt.style.use('seaborn-white')
plt.title('Execution time distribution of ACGAN-AC-RFC-Th model(quantized) running on multicore environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

plt.hist(run_in_tvm_quantified_cpu_size_4_time_data,bins=300)
```

```python
# Multicore vs Multicore(quantized)
plt.title('ACGAN-AC-RFC-Th model vs ACGAN-AC-RFC-Th model(quantized) on multicore environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

x1 = run_in_tvm_cpu_size_4_time_data
x2 = run_in_tvm_quantified_cpu_size_4_time_data
kwargs1 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Multicore")
kwargs2 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Multicore(quantized)")
plt.hist(x1, **kwargs1)
plt.hist(x2, **kwargs2)
plt.legend() # show label
```

## Overall comparison

```python
# Overall comparison
plt.title('ACGAN-AC-RFC-Th model vs ACGAN-AC-RFC-Th model(quantized) on single/multicore environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

x1 = run_in_tvm_cpu_size_0_time_data
x2 = run_in_tvm_quantified_cpu_size_0_time_data
x3 = run_in_tvm_cpu_size_4_time_data
x4 = run_in_tvm_quantified_cpu_size_4_time_data
kwargs1 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Single-core")
kwargs2 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Single-core(quantized)")
kwargs3 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Multicore")
kwargs4 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Multicore(quantized)")
plt.hist(x1, **kwargs1)
plt.hist(x2, **kwargs2)
plt.hist(x3, **kwargs3)
plt.hist(x4, **kwargs4)
plt.legend() # show label
```
