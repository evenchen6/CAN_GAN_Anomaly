## Performance Analysis Evaluation

```python
# define time data path
import numpy as np
import matplotlib.pyplot as plt

run_in_tvm_cpu_size_0_path = './ACGIDS_tvm_cpu_size_0.npz' # Single-core
run_in_tvm_quantified_cpu_size_0_path = './ACGIDS_tvm(quantified)_cpu_size_0.npz' # Single-core(quantized)
run_in_tvm_cpu_size_4_path = './ACGIDS_tvm_cpu_size_4.npz' # Multicore
run_in_tvm_quantified_cpu_size_4_path = './ACGIDS_tvm(quantified)_cpu_size_4.npz' # Multicore(quantized)

run_in_tvm_cpu_size_0_time_data = np.load(run_in_tvm_cpu_size_0_path)["time_data"]
run_in_tvm_quantified_cpu_size_0_time_data = np.load(run_in_tvm_quantified_cpu_size_0_path)["time_data"]
run_in_tvm_cpu_size_4_time_data = np.load(run_in_tvm_cpu_size_4_path)["time_data"]
run_in_tvm_quantified_cpu_size_4_time_data = np.load(run_in_tvm_quantified_cpu_size_4_path)["time_data"]
```

### Single-core vs Single-core(quantized) （cpu num is 1）

```python
# performance evaluation
# Single-core
plt.style.use('seaborn-white')
plt.title('Execution time distribution of ACGIDS model running on single core environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

plt.hist(run_in_tvm_cpu_size_0_time_data, bins=300)
```

```python
# performance evaluation
# Single-core(quantized)
plt.style.use('seaborn-white')
plt.title('Execution time distribution of ACGIDS model(quantized) running on single core environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

plt.hist(run_in_tvm_quantified_cpu_size_0_time_data,bins=300)
```

```python
# Single-core vs Single-core(quantized)
plt.title('ACGIDS model vs ACGIDS model(quantized) on single core environment')
plt.xlabel('Execution Time(ms)')
plt.ylabel('No. Samples')

x1 = run_in_tvm_cpu_size_0_time_data
x2 = run_in_tvm_quantified_cpu_size_0_time_data
kwargs1 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Single-core")
kwargs2 = dict(histtype='stepfilled', alpha=0.3, bins=300,label="Single-core(quantized)")
plt.hist(x1, **kwargs1)
plt.hist(x2, **kwargs2)
plt.legend() # show label
```
